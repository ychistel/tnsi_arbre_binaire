```python
class ABR:
    """Impl√©mentation d'un arbre binaire de recherche (ABR)"""
    def __init__(self,valeur=None):
        self.valeur = valeur
        self.fg = None
        self.fd = None
        
    def estVide(self):
        return self.valeur == None
    
    def insererElement(self,e):
        if self.estVide():
            self.valeur = e
        else:
            if e < self.valeur:
                if self.fg:
                    self.fg.insererElement(e)
                else:
                    self.fg = ABR(e)
            if e > self.valeur:
                if self.fd:
                    self.fd.insererElement(e)
                else:
                    self.fd = ABR(e)

    def appartient(self,e):
        if self.estVide():
            return False
        else:
            if e < self.valeur:
                if self.fg:
                    return self.fg.appartient(e)
                else:
                    return False
            elif e > self.valeur:
                if self.fd:
                    return self.fd.appartient(e)
                else:
                    return False
            else:
                return True
             
    def __str__(self):
        return '(' + str(self.valeur) + ',' + str(self.fg) + ',' + str(self.fd) + ')'
                    
a = ABR(8)
a.insererElement(5)
a.insererElement(12)
a.insererElement(7)
a.insererElement(3)
```
