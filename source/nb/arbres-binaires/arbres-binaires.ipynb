```python
n=Noeud(10,\
        Noeud(7,Noeud(9,Noeud(5)),Noeud(4,Noeud(3),Noeud(2))),\
        Noeud(6,Noeud(1),Noeud(8)))
a2=Arbre(n)
afficher(a2)
```

```python
def taille(a):
    if a.est_vide():
        return 0
    else:
        return 1+ taille(a.fils_gauche())+taille(a.fils_droit())
```

```python
taille(a1)
```

```python
taille(a2)
```

```python
def hauteur(a):
    if a.est_vide():
        return -1
    else:
        return 1 + max(hauteur(a.fils_gauche()),hauteur(a.fils_droit()))
```

```python
hauteur(a1)
```

```python
hauteur(a2)
```

```python
def niveau(a,p,nds):
    if p == 0:
        if not a.est_vide():
            # on empile dans une liste les valeurs
            nds.append(a.racine.valeur)
    else:
        niveau(a.fils_gauche(),p-1,nds)
        niveau(a.fils_droit(),p-1,nds)
    return nds
```

```python
arbre=a1
for i in range(hauteur(arbre)+1):
    print("Niveau %s : %s" % (i,niveau(arbre,i,[])))
```
