## Algorithme de taille

    fonction taille(arbre)
        si arbre est vide:
            renvoyer 0
        sinon:
            renvoyer 1 + taille(arbre gauche) + taille(arbre droit)

```python
from rcviz import viz

@viz
def taille(arbre):
    if arbre.est_vide():
        return 0
    else:
        return 1 + taille(arbre.fils_gauche()) + taille(arbre.fils_droit())
    
print(taille(a1))
taille.callgraph()
```

```python
taille(a2)
```

## hauteur d'un arbre



```python
def hauteur(arbre):
    if arbre.est_vide():
        return -1
    else:
        return 1 + max(hauteur(arbre.fils_gauche()),hauteur(arbre.fils_droit()))
```

```python
hauteur(a2)
```

## Niveau

```python
def niveau(arbre,p):
    if p == 0:
        if arbre.racine.valeur != None:
            return [arbre.racine.valeur]
        else:
            return []
    else:
        return niveau(arbre.fils_gauche(),p-1) + niveau(arbre.fils_droit(),p-1)
```

```python
niveau(a1,3)
```
